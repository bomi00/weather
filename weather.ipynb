```python
values = df2.mean(axis=1)
```

```python
dates = np.arange(1319)
```

```python
plt.figure(figsize=(15, 7))
plt.plot(dates, values)
plt.title('Austin Weather')
plt.grid(True)
plt.show()
```

```python
#discrete time signal
#A signal is said to be deterministic if there is no uncertainty with respect to its value at any instant of time. 
#Or, signals which can be defined exactly by a mathematical formula are known as deterministic signals.
```

```python
#discrete time signal
t = np.linspace(-1, 1,1319)
plt.plot(t, -1 * np.sin(2*np.pi*values*t));
plt.xlabel('x');
plt.ylabel('y');
plt.title('Austin Weather');
plt.xlim([-1, 0.05]);
#@savefig sineplot.png
plt.show()
```

```python
#Deterministic and Non-deterministic Signals
```

```python
vt = 0.10/1319
d = np.cos(3 * np.pi * 1319 *values * vt);
plt.xlabel('x');
plt.ylabel('y');
plt.title('Austin Weather');
plt.plot(d, values);
plt.show
```

```python
#Even signals are symmetric around vertical axis, and Odd signals are symmetric about origin. 
#Even Signal: A signal is referred to as an even if it is identical to its time-reversed counterparts
```
